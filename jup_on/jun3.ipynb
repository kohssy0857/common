```python
import numpy as np
wei=np.array([0.1,0.2,-0.1])
toes=np.array([8.5,9.5,9.9,9.0])
wlrec=np.array([0.65,0.8,0.8,0.9])
nfans=np.array([1.2,1.3,0.5,1.0])
winR=np.array([1,1,0,1])
alp=0.01
inp=np.array([toes,wlrec,nfans])
for i in range(4):
    lis=inp.T[i]
    pred=np.dot(wei,lis)
    err=pred-winR[i]
    DaM=lis*err
    if winR[i]==1:
        msg="勝ち"
    else:
        msg="負け"
    print(i+1,"回目　予測値",pred,"実際の結果",msg,"誤差",err,"各重み",wei)
    wei-=(DaM*alp)
print(wei)
#     wei=wei-DaM*lis
```
