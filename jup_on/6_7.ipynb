```python
import numpy as np
# 重み
wei=np.array([[0.1,0.1,-0.3],[0.1,0.2,0.0],[0.0,1.3,0.1]]).T
# 入力値
toes=np.array([8.5,9.5,9.9,9.0])
wlrec=np.array([0.65,0.8,0.8,0.9])
nfans=np.array([1.2,1.3,0.5,1.0])
inp=np.array([toes,wlrec,nfans]).T
# 目標値
hurt=np.array([0.1,0.0,0.0,0.1])
sad=np.array([0.1,0.0,0.1,0.2])
winR=np.array([1,1,0,1])
goal=np.array([hurt,winR,sad]).T
# 学習速度
alp=0.001
# 実行部
for j in range(100):
    for i in range(4):
        pred=np.dot(inp[i],wei)
        lis=goal[i]
        err=pred-lis
        DaM=err*inp[i]
        print(j+1,"週目",i+1,"回目　予測値",pred,"実際の結果",lis,"誤差",err,"各重み",wei)
        wei-=(DaM*alp)
print("重み",wei)
```
