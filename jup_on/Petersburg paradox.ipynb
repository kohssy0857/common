```python
import numpy as np
import matplotlib.pyplot as plt
import random
flag=True
ex_vals=np.array([])
for j in [30,100,500,1000,5000,10000,100000]:
    scores=np.array([])
    for i in range(j):
        score=1
        while flag:
            if random.random() >= 0.5 :
                score*=2
            else:
                scores=np.append(scores,score)
                break
    ex_val=np.sum(scores)/j
    print(ex_val)
    ex_vals=np.append(ex_vals,ex_val)
    plt.hist(scores,bins=30)
    plt.grid()
    plt.show()
print(ex_vals)
```
